# 01 - Data Overview and Validation

This notebook performs the mandatory data accountability checks, validates schema, and documents leakage constraints used by all downstream tasks.

```python
# Shared setup
import warnings
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
warnings.filterwarnings('ignore')

print('pandas:', pd.__version__)
print('numpy:', np.__version__)
print('seed:', SEED)
```

```python
DATA_PATH = 'Dataset .csv'
df = pd.read_csv(DATA_PATH)
print('Loaded shape:', df.shape)
```

```python
import io

print('--- HEAD ---')
display(df.head())

print('--- INFO ---')
buffer = io.StringIO()
df.info(buf=buffer)
print(buffer.getvalue())

print('--- DESCRIBE (all columns) ---')
display(df.describe(include='all').T)

print('--- SHAPE ---')
print(df.shape)

print('--- NULL COUNTS ---')
display(df.isnull().sum().sort_values(ascending=False).to_frame('null_count'))
```

```python
EXPECTED_COLUMNS = [
    'Restaurant ID', 'Restaurant Name', 'Country Code', 'City', 'Address',
    'Locality', 'Locality Verbose', 'Longitude', 'Latitude', 'Cuisines',
    'Average Cost for two', 'Currency', 'Has Table booking',
    'Has Online delivery', 'Is delivering now', 'Switch to order menu',
    'Price range', 'Aggregate rating', 'Rating color', 'Rating text', 'Votes'
]

missing = sorted(set(EXPECTED_COLUMNS) - set(df.columns))
extra = sorted(set(df.columns) - set(EXPECTED_COLUMNS))

assert not missing, f'Missing required columns: {missing}'
print('Schema assertion passed. All required columns are present.')
if extra:
    print('Additional columns detected:', extra)
```
